# 04. Line Chart

```python
#sampling data: to create a 70/30 ratio sample
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
#creating samples of df
big_samp = df[dev]
small_samp = df[~dev]
```

```python
#check
len(df)
```

```python
#check 
len(big_samp) + len(small_samp)
```

```python
#reducing sample to only necessary columns
sample = small_samp[['order_time','prices']]
```

```python
#creating line chart
line_price_by_order_time = sns.lineplot(data = sample, x = 'order_time',y = 'prices')
```

```python
#save as an image
line_price_by_order_time.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_price_by_order_time.png'))
```

### task 5 - Higher priced items tend to be purchased in the morning (around 5).

```python
#reducing sample to only necessary columns
sample2 = small_samp[['Age','dependents','marital_status']]
```

```python
#task 6 - create line chart to see connection between age and family situation
line_age_by_family = sns.lineplot(data = sample2, x = 'Age', y = 'dependents', color='#68bf43')
plt.xlabel("Age")
plt.ylabel("Average # of Dependents")
plt.title("Average # of Dependents by Instacart Customer Age")
```

```python
#save as an image
line_age_by_family.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_age_by_family.png'))
```
